```python
from sklearn.preprocessing import LabelEncoder
ob1=LabelEncoder()
ob2=LabelEncoder()
ob3=LabelEncoder()
ob4=LabelEncoder()
ob5=LabelEncoder()
data['Gender']=ob1.fit_transform(data['Gender'])
data['Occupation']=ob2.fit_transform(data['Occupation'])
data['BMI Category']=ob3.fit_transform(data['BMI Category'])
data['Blood Pressure']=ob4.fit_transform(data['Blood Pressure'])
data['Sleep Disorder']=ob5.fit_transform(data['Sleep Disorder'])
data.head()
```

```python
data.corr()
```

```python
plt.figure(figsize=(15,7))
sns.heatmap(data.select_dtypes(['number']).corr(),annot=True)
```

```python
data.drop(['Person ID'],axis=1,inplace=True)
data
```

```python
x=data.drop(['Stress Level'],axis=1)
y=data['Stress Level']
x
```

```python
from sklearn.feature_selection import SelectKBest
bestfeature=SelectKBest()
fit=bestfeature.fit(x,y)
fit
```

```python
dfscore=pd.DataFrame(fit.scores_,columns=['score'])
dfcol=pd.DataFrame(x.columns,columns=['feature'])
fs=pd.concat([dfcol,dfscore],axis=1)
fs
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=42)
xtest
```

**LOGISTIC REGRESSION**

```python
from sklearn.linear_model import LogisticRegression
model1=LogisticRegression()
model1.fit(xtrain,ytrain)
```

```python
print(model1.score(xtest,ytest))
print(model1.score(xtrain,ytrain))
```

```python
y_pred1=model1.predict(xtest)
y_pred1
```

```python
model1.predict([[1,28,5,5.9,4,30,2,11,85,3000,1]])
```
